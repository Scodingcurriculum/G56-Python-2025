Visualize and compare fractions using colorful pie charts in Python. üçïüìä. Take inputs for numerators and denominators from user.

```python
import matplotlib.pyplot as plt

# Step 1: Get input from the user for two fractions
numerator1 = int(input("Enter the numerator of the first fraction: "))
denominator1 = int(input("Enter the denominator of the first fraction: "))

numerator2 = int(input("Enter the numerator of the second fraction: "))
denominator2 = int(input("Enter the denominator of the second fraction: "))

# Step 2: Calculate the actual values for the fractions
fraction1 = numerator1 / denominator1
fraction2 = numerator2 / denominator2
remaining1 = 1 - fraction1  # Remaining part of the first fraction
remaining2 = 1 - fraction2  # Remaining part of the second fraction

# Step 3: Create Pie charts
labels = ['Fraction', 'Remaining']
colors = ['#ff9999', '#66b3ff']  # Color for fraction and remaining part

# Create Pie Chart for Fraction 1
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.pie([fraction1, remaining1], labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title(f'Fraction 1: {numerator1}/{denominator1} = {fraction1:.2f}')

# Create Pie Chart for Fraction 2
plt.subplot(1, 2, 2)
plt.pie([fraction2, remaining2], labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title(f'Fraction 2: {numerator2}/{denominator2} = {fraction2:.2f}')

# Step 4: Display the Pie charts
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Step 1: Get input from the user for two fractions
numerator1 = int(input("Enter the numerator of the first fraction: "))
denominator1 = int(input("Enter the denominator of the first fraction: "))

numerator2 = int(input("Enter the numerator of the second fraction: "))
denominator2 = int(input("Enter the denominator of the second fraction: "))

# Step 2: Calculate the actual values for the fractions
fraction1 = numerator1 / denominator1
fraction2 = numerator2 / denominator2
remaining1 = 1 - fraction1  # Remaining part of the first fraction
remaining2 = 1 - fraction2  # Remaining part of the second fraction

# Step 3: Create Pie charts
labels = ['Fraction', 'Remaining']

# Define custom colors for better visualization
colors1 = ['#ff5733', '#c2c2f0']  # Red and light purple for fraction 1

# Explode the largest slice to of the first fraction emphasize it
explode1 = (0.1, 0)

# Create Pie Chart for Fraction 1
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.pie([fraction1, remaining1], labels=labels, colors=colors1,
        autopct='%1.1f%%', startangle=90, explode=explode1, shadow=True)



colors2 = ['#33cc33', '#ffb3e6']  # Green and light pink for fraction 2


explode2 = (0.1, 0)  # Explode the 'Fraction' part for the second pie chart




plt.title(f'Fraction 1: {numerator1}/{denominator1} = {fraction1:.2f}', fontsize=14)

# Create Pie Chart for Fraction 2
plt.subplot(1, 2, 2)
plt.pie([fraction2, remaining2], labels=labels, colors=colors2, autopct='%1.1f%%', startangle=90, explode=explode2, shadow=True)
plt.title(f'Fraction 2: {numerator2}/{denominator2} = {fraction2:.2f}', fontsize=14)

# Step 4: Display the Pie charts with a tight layout
plt.tight_layout()
plt.show()
```

Create a pie chart to represent different types of fruits in a basket by taking inputs from user for different number of fruits.

```python
import matplotlib.pyplot as plt

# Step 1: Get user input for different types of fruits
apple_count = int(input("Enter the number of apples in the basket: "))
banana_count = int(input("Enter the number of bananas in the basket: "))
orange_count = int(input("Enter the number of oranges in the basket: "))
grape_count = int(input("Enter the bunches of grapes in the basket: "))
other_count = int(input("Enter the number of other fruits in the basket: "))

# Step 2: List of fruits and their respective counts
fruit_labels = ['Apples', 'Bananas', 'Oranges', 'Grapes', 'Other Fruits']
fruit_counts = [apple_count, banana_count, orange_count, grape_count, other_count]

# Step 3: Define custom colors for the fruits
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']  # Colors for each fruit

# Step 4: Create Pie chart
plt.figure(figsize=(8, 8))  # Set the size of the pie chart
plt.pie(fruit_counts, labels=fruit_labels, colors=colors, autopct='%1.1f%%', startangle=90, shadow=True)

# Step 5: Add a title
plt.title('Fruit Distribution in the Basket', fontsize=16)

# Step 6: Display the Pie chart
plt.show()
```
