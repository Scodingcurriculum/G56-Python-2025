
Store the number of gold, silver, and bronze medals won by different countries in a dictionary. Allow users to view rankings, sort by total medals, and check specific country statistics. Additionallly: Calculate total medals received by a country

```python
# Dictionary to store medal statistics for different countries
medals = {
    "USA": {"gold": 39, "silver": 41, "bronze": 33},
    "China": {"gold": 38, "silver": 32, "bronze": 18},
    "Japan": {"gold": 27, "silver": 14, "bronze": 17},
    "Germany": {"gold": 10, "silver": 11, "bronze": 16},
    "UK": {"gold": 22, "silver": 21, "bronze": 22},
    "Russia": {"gold": 19, "silver": 18, "bronze": 19}
}

def view_rankings():
    """Display the countries ranked by total medals (gold + silver + bronze)."""
    print("\nMedal Rankings (Sorted by Total Medals):\n")
    # Create a list of countries with their total medals
    total_medals = {country: sum(medals[country].values()) for country in medals}
    # Sort the countries by total medals in descending order
    sorted_medals = sorted(total_medals.items(), key=lambda x: x[1], reverse=True)

    # Display the rankings
    for rank, (country, total) in enumerate(sorted_medals, 1):
        gold = medals[country]["gold"]
        silver = medals[country]["silver"]
        bronze = medals[country]["bronze"]
        print(f"{rank}. {country}: {gold} Gold, {silver} Silver, {bronze} Bronze - Total: {total} Medals")

def check_country_stats():
    """Allow the user to check a specific country's medal statistics."""
    country = input("\nEnter the country name to check its medal statistics: ").title()
    if country in medals:
        gold = medals[country]["gold"]
        silver = medals[country]["silver"]
        bronze = medals[country]["bronze"]
        total = gold + silver + bronze
        print(f"\n{country} - Gold: {gold}, Silver: {silver}, Bronze: {bronze} - Total: {total} Medals")
    else:
        print("\n‚ùå Country not found. Please check the name and try again.")

def main():
    """Main function to interact with the user."""
    while True:
        print("\nWelcome to the Medal Rankings System!")
        print("1. View Medal Rankings")
        print("2. Check Specific Country's Stats")
        print("3. Exit")

        choice = input("\nEnter your choice (1, 2, or 3): ")

        if choice == "1":
            view_rankings()
        elif choice == "2":
            check_country_stats()
        elif choice == "3":
            print("\nExiting the system. Thank you for using the Medal Rankings System!")
            break
        else:
            print("\n‚ùå Invalid choice. Please select 1, 2, or 3.")

# Run the main program
main()
```

SuperHero Database: Store superhero names as keys and their powers as values in a dictionary. Allow users to:
1. Add a new superhero.
2. Find a superhero‚Äôs power.
3. List all superheroes.

```python
# Initialize the superhero database as an empty dictionary
superhero_database = {}

def add_superhero():
    """Function to add a new superhero and their power to the database."""
    name = input("Enter the superhero name: ").title()
    power = input(f"Enter {name}'s superpower: ").title()
    superhero_database[name] = power
    print(f"\n‚úÖ {name} has been added to the superhero database with the power '{power}'!")

def find_superhero_power():
    """Function to find a superhero's power by their name."""
    name = input("Enter the superhero name to find their power: ").title()
    if name in superhero_database:
        print(f"\nüîç {name}'s superpower is: {superhero_database[name]}")
    else:
        print(f"\n‚ùå Sorry, {name} is not in the superhero database.")

def list_all_superheroes():
    """Function to list all superheroes and their powers."""
    if superhero_database:
        print("\nList of All Superheroes and Their Powers:")
        for name, power in superhero_database.items():
            print(f"- {name}: {power}")
    else:
        print("\n‚ùå No superheroes have been added to the database yet.")

def main():
    """Main function to run the superhero database program."""
    while True:
        print("\nWelcome to the Superhero Database!")
        print("1. Add a new superhero")
        print("2. Find a superhero's power")
        print("3. List all superheroes")
        print("4. Exit")

        choice = input("\nEnter your choice (1, 2, 3, or 4): ")

        if choice == "1":
            add_superhero()
        elif choice == "2":
            find_superhero_power()
        elif choice == "3":
            list_all_superheroes()
        elif choice == "4":
            print("\nExiting the Superhero Database. Thank you for using it!")
            break
        else:
            print("\n‚ùå Invalid choice. Please select 1, 2, 3, or 4.")

# Run the main program
main()
```
